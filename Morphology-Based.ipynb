```python
# Import libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
 
image = cv2.imread('inputs/P3/cart_1.png')
image = image[:][55:-55]

gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
binary = np.zeros(gray.shape)
for i in range(gray.shape[0]):
    for j in range(gray.shape[1]):
        if image[i][j][0] == image[i][j][1] and image[i][j][2] == image[i][j][1] and image[i][j][1]<190 and image[i][j][1]>25:
            binary[i][j] = 0
        else :
            binary[i][j] = 255


plt.figure(figsize=(25,15))
plt.imshow(binary,cmap="gray")
plt.show()
```

```python
image_blur = cv2.medianBlur(np.uint8(binary),37)
plt.figure(figsize=(25,15))
plt.imshow(image_blur,cmap="gray")
plt.show()
```

```python
kernel = np.ones((5, 5), np.float32)
image_dilated = cv2.morphologyEx(image_blur, cv2.MORPH_DILATE, kernel, iterations=10)
plt.figure(figsize=(25,15))
plt.imshow(image_dilated,cmap="gray")
plt.show()
```

```python
(cnt, hierarchy) = cv2.findContours(
    image_dilated.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
cv2.drawContours(rgb, cnt, -1, (0, 255, 0), 2)
 
plt.figure(figsize=(25,15))
plt.imshow(rgb,cmap="gray")
plt.show()
print("items in the image : ", len(cnt))
```

```python
counts = []
count = len(cnt)
j = 0
image_erod = image_dilated
kernel = np.ones((5, 5), np.float32)
while count !=0:
    image_erod = cv2.morphologyEx(image_erod, cv2.MORPH_ERODE, kernel, iterations=1)
    (cnt, hierarchy) = cv2.findContours(image_erod.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
    if j%15 ==0:
      plt.imshow(image_erod,cmap="gray")
      plt.show()
    count = len(cnt)
    counts.append(count)
    j+=1
    print("count : ", count)
    print("iterations : ", j)
    print("-----------------------------------")
```
